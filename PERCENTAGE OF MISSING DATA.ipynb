```python
import pandas as pd
```

```python


# Load your Excel file
#file_path = "path_to_your_file.xlsx"
data = pd.read_excel('Relative_Humidity_with_all_months.xlsx')
data
```

```python
#Count total cells
total_cells = data.size

# Count non-missing cells
non_missing_cells = data.count().sum()

# Calculate percentage of available data
percentage_available = (non_missing_cells / total_cells) * 100

print(f"Percentage of available data: {percentage_available:.2f}%")
```
